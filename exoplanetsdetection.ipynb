```python
def mean_normalise(series):
  norm_Series=(series-series.mean())/(series.max()-series.min())
  return norm_Series
```

```python
norm_train_df = exo_train_df.iloc[:, 1:].apply(mean_normalise, axis=1)
norm_train_df.head()
```

```python
exo_train_df.head()
```

```python
norm_train_df.insert(loc=0, column='LABEL', value=exo_train_df['LABEL'])
norm_train_df.head()
```

```python
norm_test_df=exo_test_df.iloc[:,1:].apply(mean_normalise,axis=1)
norm_test_df.head()
```

```python
norm_test_df.insert(loc=0,column="LABEL",value=exo_test_df['LABEL'])
norm_test_df.head()
```

```python
#creating a line plot before applying fastfourier transformation
plt.figure(figsize=(20, 5))
plt.plot(np.arange( 3198), star_0)
plt.grid() 
```

```python
fft_star_0 = np.abs(np.fft.fft(star_0))
fft_star_0
```

```python
freq=np.fft.fftfreq(len(star_0))
freq
```

```python
#creating a line plot after applying fastfourier transformation
plt.figure(figsize=(20,5))
plt.plot(freq,fft_star_0)
plt.show()
```

```python
def fast_fourier_transformation(star):
  fft_star=np.abs(np.fft.fft(star,n=len(star)))
  return fft_star
```

```python
x_fft_train_T = norm_train_df.iloc[:, 1:].T.apply(fast_fourier_transformation,axis=0)
x_fft_train = x_fft_train_T.T
x_fft_train.head()
```

```python
#applying fastfourier transformation to test data
x_fft_test_T=norm_test_df.iloc[:,1:].T.apply(fast_fourier_transformation,axis=0)
x_fft_test=x_fft_test_T.T
x_fft_test.head()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy=1)
x_fft_train_res, y_fft_train_res = smote.fit_resample(x_fft_train ,y_train)
```

```python
print(type(x_fft_train_res),x_fft_train_res.shape)
print(type(y_fft_train_res),y_fft_train_res.shape)
```

```python
print(sum(y_fft_train_res==1))
print(sum(y_fft_train_res==2))
print(y_fft_train_res.value_counts())
```

```python
ra=RandomForestClassifier(n_estimators=50,n_jobs=-1)
ra.fit(x_fft_train_res,y_fft_train_res)
predict=ra.predict(x_fft_test)
pd.Series(predict).value_counts()
```

```python
confusion_matrix(y_test,predict)
```

```python
print(classification_report(y_test,predict))
```
